
* Hardware
Definition: Physical parts of a computer that you can see and touch
Examples: Keyboard, mouse, monitor, CPU, hard disk, printer
Nature: Tangible (physical objects)
Function: Performs tasks when instructed
* Software
Definition: Programs or instructions that tell the hardware what to do
Examples: Operating systems like Windows, apps like Chrome, games, MS Word
Nature: Intangible (you cannot touch it)
Function: Controls and operates the hardware
